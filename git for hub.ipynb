```python
import pandas as pd

data = {
    "Month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun",
              "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"],
    "Sales": [1200, 1350, 1500, 1600, 1700, 1650,
              1800, 1900, 1850, 2000, 2100, 2300]
}

df = pd.DataFrame(data)
print(df)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
plt.plot(df["Month"], df["Sales"], marker='o', color='darkblue', linewidth=2)

plt.title("Monthly Sales Trend", fontsize=16)
plt.xlabel("Month")
plt.ylabel("Sales ($)")
plt.grid(True)

plt.show()
```

```python
import pandas as pd
import numpy as np

np.random.seed(10)

departments = ["HR", "Finance", "IT", "Marketing"]
salaries = {
    "Department": np.random.choice(departments, 50),
    "Salary": np.random.randint(30000, 100000, 50)
}

df = pd.DataFrame(salaries)
print(df.head())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
sns.boxplot(x="Department", y="Salary", data=df, palette="Set2")
plt.title("Box Plot of Salary by Department")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.violinplot(x="Department", y="Salary", data=df, palette="Set3")
plt.title("Violin Plot of Salary by Department")
plt.show()
```
